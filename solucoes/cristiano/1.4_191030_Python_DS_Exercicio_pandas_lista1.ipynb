### 4 mostre o valor mais alto e mais baixo de horsepower


```python
print(from_url.describe()["horsepower"])

print("\n")

display("Maximum Horsepower", from_url[from_url['horsepower']==from_url['horsepower'].max()])
display("Minimum Horsepower", from_url[from_url['horsepower']==from_url['horsepower'].min()])


display("MAXIMUM HORSEPOWER:", from_url[['company', 'horsepower']][from_url.horsepower == from_url.horsepower.max()])

display("MINIMUM HORSEPOWER:", from_url[['company', 'horsepower']][from_url.horsepower == from_url.horsepower.min()])
```
