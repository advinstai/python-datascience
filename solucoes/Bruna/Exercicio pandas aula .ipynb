### Exemplo utilizando a biblioteca pandas (aula)

Abrir o arquivo Auto.csv    
Contar a quantidade de elementos nulos por coluna    
Criar um dataram eliminando as linhas com valores nulos dataframe   
Criar um dataframe transformando os valores nulos para zero     
Criar um dataframe trocando nulo por média da coluna    

```python
import pandas as pd
import numpy as np

#abrindo o arquivo por html
url='https://raw.githubusercontent.com/advinstai/python-datascience/master/csv/Auto.csv'
from_url = pd.read_table(url, sep=',')  #le a tabela e separa por virgula


from_url.describe()

#Contar a quantidade de elementos nulos por coluna
valor = from_url.replace("?",np.nan)
valor.isnull().sum()

#Criar um dataram eliminando as linhas com valores nulos dataframe
dfnulo = valor.dropna()
dfnulo.isnull().sum()

#Criar um dataframe transformando os valores nulos para zero
dfzero = valor.replace(np.nan,0)
dfzero.loc[dfzero["horsepower"]==0]

#Criar um dataframe trocando nulo por média da coluna 
dfzero["horsepower"] = pd.to_numeric(dfzero["horsepower"]).astype(int)

media = dfzero["horsepower"].mean()

#Criar um dataframe transformando os valores nulos para zero
media2 = dfzero.loc[dfzero["horsepower"]!=0, "horsepower"].mean() 


#Criar um dataframe trocando nulo por média da coluna 
dfmedia = valor.replace(np.nan,media2)

dfmedia

#dfmedia.loc[dfmedia["horsepower"] == media2]
```
