# 1 Do arquivo carros, imprima as primeiras e as últimas cinco linhas. Faça uma versão que abre o arquivo a partir do csv e outra que abre a partir da url.

```python
import pandas as pd
import numpy as np
#######################################################
# Abrir csv do arquivo no CP
carros = pd.read_csv("/home/gabrielmartins/carros.csv")
display(carros.head(5))
display(carros.tail(5))

########################################################
# Abrir csv através de uma url
url = "https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/carros.csv"
carros_url = pd.read_table(url,sep=",")
display(carros_url.head(5))
display(carros_url.tail(5))
```

# 2 Encontre o nome da empresa de carros mais cara

```python
carro_caro = carros.loc[carros["price"] == max(carros["price"])]
display(carro_caro["company"])
```

# 3 mostre a média de preços

```python
print("Média dos preços com numpy: ",np.mean(carros["price"]),"\n")
print("Média dos preços com pandas: ",carros["price"].mean())
```

# 4 mostre o valor mais alto e mais baixo de horsepower

```python
print("Mínimo da coluna horsepower: ",carros["horsepower"].min(),"\n")
print("Mínimo da coluna horsepower: ",carros["horsepower"].max())
```

# 5 mostre a informação de todos os carros da marcar toyota

```python
toyota = carros.loc[carros["company"]=="toyota"]
display(toyota)
```

# 6 faça uma função que filtra os carros por cilindro e salva em outro arquivo CSV

```python
def Cilindros(cilindros):
    num_of_cilinders = carros.loc[carros["num-of-cylinders"]==cilindros]
    num_of_cilinders.to_csv("{} cilindros.csv".format(cilindros),index=False)
    return num_of_cilinders

Cilindros("four")
```
