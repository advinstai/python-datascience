* Abrir o arquivo Auto.csv
* Contar a quantidade de elementos nulos por coluna
* Criar um dataframe eliminando as linhas com valores nulos

```python
import pandas as pd
import numpy as np

Auto1 = pd.read_csv("/home/gabrielmartins/Auto.csv", na_values ="?")

print("Quantidade de elementos nulos por coluna:","\n")
print("----------------------------")
print(Auto1.isna().sum(),"\n")
print(Auto1.isnull().sum(),"\n")
print("Tamanho:",Auto1.shape)
print("----------------------------","\n")

#########################################################
# Novo dataframe sem os valores nulos
print("Novo dataframe eliminando valores nulos")
A = Auto1.dropna().copy()
display(A)
print("Quantidade de elementos nulos por coluna:","\n")
print(A.isnull().sum(),"\n")
print("Tamanho:",A.shape)
print("----------------------------")
```
