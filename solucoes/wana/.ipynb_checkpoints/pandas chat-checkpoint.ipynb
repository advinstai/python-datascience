```python
# Abrir o arquivo Auto.csv 
import pandas as pd
import numpy as np
url = 'https://raw.githubusercontent.com/advinstai/python-datascience/master/csv/Auto.csv'

df = pd.read_table(url, sep=',', na_values='?')
df['horsepower'].isnull()
df.head()
```

```python
# Contar a quantidade de elementos nulos por coluna
df1 = df.replace('?', np.nan)

nu = df1.isnull().sum(axis = 0)
nu

m = df1['horsepower'].mean()
print(m)
```

```python
#criar um dataframe eliminando as linhas com valores nulos 

df2 = df.dropna()

df2
nu2 = df2.isnull().sum(axis = 0)
nu2
```

```python
# Criar um dataframe transformando os valores nulos para zero 

df3 = df.replace(np.nan, 0)

zeros = (df3 == 0).astype(int).sum(axis=0)

#zeros = df3[df3 == 0].values.sum()
zeros
#df3['horsepower'].values
```

```python
#criar um dataframe trocando nulo por média da coluna 

mean_z = df1['horsepower'].mean()
print(mean_z)
df3.loc[df3['horsepower'] == 0, 'horsepower']= mean_z

print(df3.loc[df3['horsepower'] == 103.15365239294711])
for i in df3['horsepower']:
    print(i)
```
