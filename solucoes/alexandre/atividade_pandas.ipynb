## Encontre o nome da empresa de carros mais cara.

```python
most_expensive = df['company'].loc[df['price']==max(df['price'])].values[0]
print("Empresa com carros mais caros: ", most_expensive)
```

## Mostre a média de preços.

```python
filtered = df[df['price']!='']
media = np.mean(filtered['price'])
print('Média dos preços: ', media)
```

## Mostre o valor mais alto e mais baixo de horsepower.

```python
maximo = np.max(df['horsepower'])
minimo = np.min(df['horsepower'])
print("Máximo de cavalos: ", maximo)
print("Mínimo de cavalos: ", minimo)
```

## Mostre a informação de todos os carros da marcar toyota.

```python
df_toyota = df[df['company']=='toyota']
display(df_toyota)
```
